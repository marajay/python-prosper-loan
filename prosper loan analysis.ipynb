```python
loanstatus_counts = loan['loanstatus'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
loanstatus_counts.plot(kind='bar')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.title('Loan Status Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=loan, x='loanstatus', y='borrowerrate')
plt.xlabel('loanstatus')
plt.ylabel('borrowerrate')
plt.title('loan status vs. borrowerrate')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data = loan, x='term',color='Red');
plt.title('loantermdistribution')
```
